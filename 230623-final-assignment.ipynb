```python
from sklearn.datasets import make_blobs 
X, _=make_blobs(n_samples=100, centers=5, n_features=2, random_state=0)
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering 
from sklearn.cluster import DBSCAN
```

```python
i=0
while(i!=4):
    print('[ Student ID: 2011429 ]')
    print('[ Name: 김윤서 ]\n')
    print('1. K-means')
    print('2. Hierarchical clustering')
    print('3. DBSCAN')
    print('4. Quit')
    i=int(input('>'))
    
    if(i==1):
        n=int(input('n_clusters: '))
        r=int(input('random_state: '))
        print('<Clustering result>')
        kmeans=KMeans(n_clusters=n, random_state=r, n_init="auto").fit(X)
        #print(kmeans.cluster_centers_)
        print(kmeans.labels_)
        #print(kmeans.inertia_)
        print('\n')
        
    elif(i==2):
        n=int(input('n_clusters: '))
        l=input('linkage: ')
        print('<Clustering result>')
        Clustering=AgglomerativeClustering(n_clusters=n, linkage=l).fit(X)
        #print(Clustering.n_clusters_)
        print(Clustering.labels_)
        #print(Clustering.n_leaves_)
        #print(Clustering.children_)
        print('\n')
        
    elif(i==3):
        e=float(input('eps: '))
        m=int(input('min_samples: '))
        print('<Clustering result>')
        clustering=DBSCAN(eps=e, min_samples=m).fit(X)
        #print(clustering.core_sample_indices_)
        print(clustering.labels_)
        print('\n')
```
